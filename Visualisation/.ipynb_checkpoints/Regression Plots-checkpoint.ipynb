```python
# The normal imports
import numpy as np
from numpy.random import randn
import pandas as pd

# Import the stats library from numpy
from scipy import stats

# These are the plotting modules adn libraries we'll use:
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

# Command so that plots appear in the iPython Notebook
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.lmplot('total_bill','tip', tips)
```

```python
sns.lmplot('total_bill','tip',tips, scatter_kws ={'marker':'o', 'color':'indianred'},
            line_kws={'linewidth':1, 'color':'blue'})
```

```python
sns.lmplot('total_bill','tip',tips,order=4 ,scatter_kws ={'marker':'o', 'color':'indianred'},
            line_kws={'linewidth':1, 'color':'blue'})
```

```python
sns.lmplot('total_bill','tip',tips, fit_reg=False)
```

```python
# lmplot() also works on discrete variables, such as the percentage of the tip

# Create a new column for tip percentage
tips['tip_per'] = 100 * tips['tip']/ tips['total_bill']
```

```python
tips.head()
```

```python
sns.lmplot('size','tip_per',tips)
```

```python
#We can also add jitter to this

sns.lmplot('size','tip_per', tips, x_jitter = .1)
```

```python
# We can also estimate the tendency of each bin (size of party in this case)

sns.lmplot('size','tip_per', tips, x_estimator=np.mean)
```

```python
# Interesting, looks like there is more variance for party sizes of 1 then 2-4
```

```python
# We can use the hue facet to automatically define subsets along a column

# Plot, note the markers argument

sns.lmplot('total_bill','tip_per',tips,hue='sex', markers=["x","o"])
```

```python
# Does day make a difference?
sns.lmplot("total_bill", "tip_per", tips, hue="day")
```

```python
# Finally it should be noted that Seabron supports LOESS model fitting
url = 'http://en.wikipedia.org/wiki/Local_regression'

sns.lmplot("total_bill", "tip_per", tips, lowess=True, line_kws={"color": 'black'});
```

```python
# The lmplot() we've been using is actually using a lower-level function, regplot()

sns.regplot("total_bill","tip_per",tips)
```

```python
# reg_plot can be added to existing axes without modifying anything in the figure

# Create figure with 2 subplots
fig, (axis1,axis2) = plt.subplots(1,2,sharey =True)

sns.regplot("total_bill","tip_per",tips,ax=axis1)
sns.violinplot(tips['tip_per'],tips['size'],color='reds_r',ax=axis2)
```
